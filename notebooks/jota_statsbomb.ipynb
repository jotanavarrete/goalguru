```python

for i, row in competitions.iterrows():
    with open(f"../raw_data/open-data/data/matches/{row['competition_id']}/{row['season_id']}.json") as file:
        json_file = json.load(file)
    
    print(row['competition_id'], row['season_id'], row['competition_name'], row['season_name'], len(json_file))
    competitions.loc[i, 'match_count'] = len(json_file)
```

```python
sns.barplot(data=competitions, y='competition_name', x='match_count');
```

```python
feats = ['pass', 'carry', 'under_pressure', 'interception', 'dribble', 'counterpress', 'duel', 'ball_receipt', 'ball_recovery', 'out', 'clearance', 'shot', 'goalkeeper', 'foul_won', 'foul_committed', 'substitution', 'block', 'bad_behaviour']
for feat in feats:
    print(feat, evs1[~evs1[feat].isna()]['player'].iloc[0], evs1[~evs1[feat].isna()]['location'].iloc[0], evs1[~evs1[feat].isna()][feat].iloc[0])
    print()
```

```python
for _, row in evs1[~evs1['tactics'].isna()].iterrows():
    print(row['period'], row['minute'], json.dumps(row['tactics'], indent=2))
    print()
```
