```python
def get_competitions(data, df_name):
    # Group the DataFrame by 'competition_id' and 'name' columns and aggregate them
    grouped_data = data.groupby(['competition_id', 'name']).size().reset_index(name='count')

    # Convert the grouped DataFrame to a list of dictionaries
    competition_list = [{'competition_id': row['competition_id'], 'name': row['name'], 'dataframe': df_name} for _, row in grouped_data.iterrows()]

    return competition_list

get_competitions(data_soccer, df_name='data_soccer')
```

```python
def get_seasons(competition_id):
    
    ## Insert dataframe
    data=data_stats
    
    # Filter the DataFrame based on the given competition_id
    filtered_data = data[data['competition_id'] == competition_id]
    
  # Check if there is data for the given competition_id
    if filtered_data.empty:
        return print("No data for the provided competition_id.")   
    
    # Get unique season IDs for the selected competition
    unique_seasons = filtered_data['season_id'].unique()

    # Create a list of dictionaries for each unique season
    seasons_list = [
        {
            'season_id': season_id,
            'season': filtered_data[filtered_data['season_id'] == season_id]['season'].iloc[0],
            'matchweeks': list(range(len(filtered_data[filtered_data['season_id'] == season_id])))
        }
        for season_id in unique_seasons
    ]

    return seasons_list

get_seasons(competition_id=0)
 
```

```python
def get_seasons(competition_id, df_name):
    # Assuming df_name is a string that identifies the dataset
    
    # Define a dictionary to map dataset names to human-readable names
    dataset_names = {
        'data_soccer': 'data_soccer',
        'data_stats': 'data_stats',  # Add other dataset names as needed
        'other_dataset': 'other',
    }
    
    # Insert the appropriate dataframe based on df_name
    if df_name == 'data_soccer':
        data = data_soccer
    elif df_name == 'data_stats':
        data = data_stats
    else:
        return print("Invalid dataset name.")
    
    # Filter the DataFrame based on the given competition_id
    filtered_data = data[data['competition_id'] == competition_id]
    
    # Check if there is data for the given competition_id
    if filtered_data.empty:
        return print("No data for the provided competition_id.")
    
    # Get unique season IDs for the selected competition
    unique_seasons = filtered_data['season_id'].unique()

    # Create a list of dictionaries for each unique season with dataset information
    seasons_list = [
        {
            'season_id': season_id,
            'season': filtered_data[filtered_data['season_id'] == season_id]['season'].iloc[0],
            'matchweeks': list(range(1, len(filtered_data[filtered_data['season_id'] == season_id]) + 1)),
            'dataset': dataset_names.get(df_name, 'Unknown')  # Get the dataset name from the mapping
        }
        for season_id in unique_seasons
    ]

    return seasons_list

get_seasons(competition_id=2, df_name='data_stats')
```

```python
def get_all_competitions():

    """
    This Fuctions Returns a list of all comptetions, according to each dataset.
    """
        
    list_copmetitions_statbombs = get_seasons(competition_id=0, df_name='data_soccer') #get_statbombs_data() ## funcion que trae todas las competencias del dataset en una lista
    list_copmetitions_soccermatch = get_seasons(competition_id=0, df_name='data_stats')   #get_soccermatch_data()  ## funcion que trae todas las competencias del dataset en una lista
    
    # Merge the two lists and remove duplicates based on the 'name' key
    merged_list = list_copmetitions_statbombs + list_copmetitions_soccermatch
          
    return merged_list

get_all_competitions()
```
